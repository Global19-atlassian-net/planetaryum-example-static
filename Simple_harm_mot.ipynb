This Jupyter notebook allows you to generate plots for the simple harmonic motion of an ideal spring.

It uses interactive sliders for the time and spring mass and stiffness. If you prefer, you can change the sliders for 'text boxes'. 

```python
%matplotlib inline 
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets
from IPython.display import display

# Sliders: the plots will only udpdate with mouse release
k = widgets.FloatSlider(x=(0.0,100.0,0.5), 
                         value=0.1, description='k',
                         continuous_update=False )
m =  widgets.FloatSlider(x=(0.0,20.0,0.5), 
                          value=0.1, description='m',
                          continuous_update=False)
                            
# In case you prefer using textboxes
#k = widgets.FloatText(value=0.1, description='k')
#m = widgets.FloatText(value=0.1, description='m')

def plot_shm(t,k,m):
    time= np.linspace(0,t,500)
    omega = np.sqrt(k/m)
    A, phi = 2,2
    
    period = 2.0 *np.pi/omega
    freq = 1.0/ period
    
    # Calculating the displacement, velocity, and acceleration
    y = A*np.sin(omega*time)
    vel = omega*A*np.cos(omega*time+phi)
    acc = -omega**2*A*np.sin(omega*time)
    
    # Plotting: 3 plots sharing the x-axis
    fig, ax=plt.subplots(3, sharex=True, figsize=(8,14))
    fig.subplots_adjust(hspace=0.10)
    
    # Plotting and personal customisation 
    col = sns.color_palette("husl",4)
    plt.suptitle('Motion of an ideal spring', 
                 fontsize=14, y=0.925)
    
    ax[0].plot(time,y, color= col[0]) 
    ax[1].plot(time,vel, color= col[1])
    ax[2].plot(time,acc, color= col[2])
    
    ax[0].set_ylabel('Displacement [m]', fontsize=12)
    ax[1].set_ylabel('Velocity [m/s]', fontsize=12)
    ax[2].set_ylabel('Acceleration[$m/s^2$]',fontsize=12)
    ax[2].set_xlabel('Time [s]', fontsize=12)
    
    
p = widgets.interactive(plot_shm, t=10, k=k, m=m, 
                        continuous_update=False)
display(p)
```
