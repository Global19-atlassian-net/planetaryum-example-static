# Polar plots of sin(k*theta)

## Author: MikeCroucher

This demonstration shows how to plot [polar functions](https://en.wikipedia.org/wiki/Polar_coordinate_system) in [matplotlib](http://matplotlib.org/) using $r=sin(k\theta)$ as an example. Even though this function is so simple, the results can be quite pretty.

```python
# Insert code here
# It should be completely self contained -- the only cell that needs to be executed to run the interaction
# The notebook should be uploaded to GitHub with this cell evaluated
# The interaction should have have an interesting set of defaults
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import arange,sin,pi
from ipywidgets import interactive

theta = arange(0, 50*pi, .01)

def sinf(k,theta):
    return sin(k*theta)

def polar_plot(k=2.65):
    ax = plt.subplot(111,projection='polar')
    ax.plot(theta, sinf(k,theta))
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(ymax=1.2)


interactive(polar_plot,k=(1,10.,0.01))
```
