# Sonic Boom

The visualization of an object  moving at speeds  around the speed of sound. The object (red point) has a position $x(t),y(t)$ and is emmiting sound at frequency `freq`, which is shown as cicles with growing radii. 

Circles (wavefronts are drawn with small opacity, this  the darker places correspond to places where more wavefronts meet and resulting sound amplitude is larger. 

One can interact with time, the speed of motion and  choose the tracjectory.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider
from IPython.core.display import display

plt.interactive(False)
f,ax = plt.subplots(1)
ax.set_aspect('equal')

@interact(t=FloatSlider(min=0.1,max=5,step=0.01,width='500px',value=3),
         v=FloatSlider(min=0.1,max=3.0,step=0.1,width='500px',value=1.0),c=["line","circle"])
def hiper(t,v,c):

    if c=="line":
        x = lambda t: v*t 
        y = lambda t: 0.*t+1.0
    if c=="circle":
        x = lambda t: 2+a*np.cos(v/a*t)
        y = lambda t: 1+a*np.sin(v/a*t)    
    freq,c,a = 22.2, 1, 1
    ts = np.arange(0.1,t,1/freq)
    ax.cla()
    for t0 in ts:
        ax.add_artist(plt.Circle( (x(t0),y(t0)),c*(t-t0),fill=False,alpha=0.12))
    ts = np.linspace(0,2*np.pi/(v/a),100)
    ax.plot(x(ts),y(ts))
    ax.plot(x(t),y(t),'ro')
    ax.set_xlim(0,4)
    ax.set_ylim(-1,3)
    display(f)
```
